```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
df = pd.read_csv("../data/df_stats.csv")
```

```python
df.head()
```

## Distribution of slope of trend fitting
- split by type (all types should be equal)

```python
fig, ax = plt.subplots(figsize=(13,9))
sns.histplot(df, ax=ax, x='m',
            hue='type', log_scale=True);
```

## Distribution of the Mean
- split by type (should be equal for all types)

```python
fig, ax = plt.subplots(figsize=(13,9))
sns.histplot(df, x='mean', ax=ax, hue='type', log_scale=True);
```

### Showing 99th Quantile

```python
df_q99 = df[df['mean']<df['mean'].quantile(.99)]
fig, ax = plt.subplots(figsize=(13,9))
sns.histplot(df_q99, x='mean', ax=ax, hue='type');
```

## Median Distribution

```python
fig, ax = plt.subplots(figsize=(13,9))
sns.histplot(df, x='q50', ax=ax, hue='type', log_scale=True);
```

```python
df_q99 = df[df['q50']<df['q50'].quantile(.99)]
fig, ax = plt.subplots(figsize=(13,9))
sns.histplot(df_q99, x='q50', ax=ax, hue='type');
```

## Standard Deviation Distribution

```python
fig, ax = plt.subplots(figsize=(13,9))
sns.histplot(df, x='std', ax=ax, hue='type', log_scale=True);
```
