```python
sample = list(df['V1'].sample(2).values)
ar1 = np.array(df[df['V1']==sample[0]].iloc[:,1:].dropna(axis=1))[0]
ar2 = np.array(df[df['V1']==sample[1]].iloc[:,1:].dropna(axis=1))[0]
aln = dtw(ar1, ar2)
aln.plot()
dtwPlotTwoWay(aln, ar1, ar2)
```

```python
ax = dtwPlotTwoWay(aln, ar1, ar2)
```

```python
fig_3way = dtwPlotThreeWay(aln, ar1, ar2)
```

```python
ax.figure.savefig("../img/dtw_ex_plain.png")
```
