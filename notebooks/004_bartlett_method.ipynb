```python
ar = np.random.rand(99)
n = ar.shape[0]
k = 4 # number of segments
l = int(np.floor(n/k))
x = np.linspace(1,n,n)
print(l)
```

```python
sns.set(font_scale=1)

fig = plt.figure(constrained_layout=True, figsize=(13,10))
gs = GridSpec(2,3, figure=fig)

ax1 = fig.add_subplot(gs[0, :])
ax2 = fig.add_subplot(gs[1, 0])
ax3 = fig.add_subplot(gs[1, 1])
ax4 = fig.add_subplot(gs[1, 2])
sns.lineplot(x=x, y=ar, label=ts_name, ax=ax1)
ylab_ax1 = r"$\displaystyle x(n)$"
ylab_ax2 =  r"$\displaystyle x_1(n)$"
ylab_ax3 =  r"$\displaystyle x_2(n)$"
ylab_ax4 =  r"$\displaystyle x_3(n)$"
ax1.set_xlabel(ylab_ax1)
sns.lineplot(x=x[:l], y=ar[:l], label=ts_name+" - Segment 1", ax=ax2)
ax2.set_xlabel(ylab_ax2)
sns.lineplot(x=x[l:2*l], y=ar[l:2*l], label=ts_name+" - Segment 2", ax=ax3)
ax3.set_xlabel(ylab_ax3)
sns.lineplot(x=x[2*l:], y=ar[2*l:], label=ts_name+" - Segment 3", ax=ax4)
ax4.set_xlabel(ylab_ax4)
plt.savefig("../img/bartlett_example.png")
```
