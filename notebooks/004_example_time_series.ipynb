```python
def generate_plots(samples: list,
                   df_all: pd.DataFrame,
                  axs ) -> None:
    for ts_name in samples:
        f = np.array(df_all[df_all['V1']==ts_name].iloc[:,1:].dropna(axis=1))[0]
        n = f.shape[0]
        
        # FFT
        fhat = np.fft.fft(f)
        PSD = np.real(fhat * np.conj(fhat) / n)
        freq = np.arange(n)/f.size
        L = np.arange(1,np.floor(n/2),dtype='int')
        
        # Hamming
        ham_window = np.hamming(f.size)
        f_ham = ham_window * f
        fh_ham = np.fft.fft(f_ham)
        PSD_ham = np.real(fhat * np.conj(fhat) / n)
        freq_ham = np.arange(n)/f.size
        L_ham = np.arange(1,np.floor(n/2),dtype='int')
        
        # Welch
        seg_length = np.floor(1/20*n)
        if seg_length == 0:
            seg_length=10
        welch_freqs, PSD_welch = signal.welch(f, nperseg=seg_length,
                                              window='hamming')
        
        axs.flatten()[0].plot(freq[L],PSD[L],lw=2)
        axs.flatten()[0].axhline(y=100000, color='r', lw=5, linestyle='--')
        axs.flatten()[1].plot(freq_ham[L],PSD_ham[L], lw=2)
        axs.flatten()[1].axhline(y=100000, color='r', lw=5, linestyle='--')
        axs.flatten()[2].plot(welch_freqs,PSD_welch, lw=2)
        axs.flatten()[2].axhline(y=20000, color='r', lw=5, linestyle='--')
        plt.setp(axs.flatten()[0].get_xticklabels(), visible=False);
        plt.setp(axs.flatten()[0].get_yticklabels(), visible=False);
        plt.setp(axs.flatten()[1].get_xticklabels(), visible=False);
        plt.setp(axs.flatten()[1].get_yticklabels(), visible=False);
        plt.setp(axs.flatten()[2].get_xticklabels(), visible=False);
        plt.setp(axs.flatten()[2].get_yticklabels(), visible=False);
#        ax.set_xscale('log')
```

```python
def generate_hists(samples: list,
                   df_all: pd.DataFrame,
                  axs ) -> None:
    for ts_name in samples:
        f = np.array(df_all[df_all['V1']==ts_name].iloc[:,1:].dropna(axis=1))[0]
        n = f.shape[0]
        x = np.linspace(1,n,n)

        
        # Mean
        mean = f.mean()
        axs.flatten()[0].plot(f)
        axs.flatten()[0].axhline(mean, color='r', lw=4)
        axs.flatten()[0].set_title("Mean")
        
        # Median
        median = np.median(f)
        axs.flatten()[1].plot(f)
        axs.flatten()[1].axhline(median, color='r', lw=4)
        axs.flatten()[1].set_title("Median")
        
        # Std
        std = np.std(f)
        y_zero = np.zeros(f.size)
        axs.flatten()[2].scatter(f,y_zero)
        axs.flatten()[2].axvline(mean-std, color='r', lw=2)
        axs.flatten()[2].axvline(mean, color='b')
        axs.flatten()[2].axvline(mean+std, color='r', lw=2)
        axs.flatten()[2].set_title("Standard Deviation")
        axs.flatten()[2].text(mean-std+50, 0.04,
                              r"$\displaystyle \mu - \sigma$", color='r',
                             fontsize=25)
        axs.flatten()[2].text(mean+50, 0.04,
                              r"$\displaystyle \mu$", 
                              fontsize=25, color='b')
        axs.flatten()[2].text(mean+std+50, 0.04,
                              r"$\displaystyle \mu + \sigma$", color='r',
                             fontsize=25)
        
        # Boxplot
        sns.boxplot(x=f, ax=axs.flatten()[3])
        axs.flatten()[3].set_title("Quantiles")
        
        # Trend
        lin_model = np.polyfit(x,f,1)
        lin_mod_fn = np.poly1d(lin_model)
        axs.flatten()[4].plot(x,lin_mod_fn(x))
        axs.flatten()[4].set_title("Trend")
        
        # Min Max
        min_val = f.min()
        max_val = f.max()
        axs.flatten()[5].plot(f)
        axs.flatten()[5].axhline(min_val, color='r')
        axs.flatten()[5].axhline(max_val, color='r')
        axs.flatten()[5].set_title("Min - Max")
        axs.flatten()[5].text(0, 3400,
                              r"Max", color='r',
                             fontsize=25)
        axs.flatten()[5].text(80, 300,
                              r"Min", color='r',
                             fontsize=25)
        
        
#         plt.setp(axs.flatten()[0].get_xticklabels(), visible=False);
#         plt.setp(axs.flatten()[0].get_yticklabels(), visible=False);
#         plt.setp(axs.flatten()[1].get_xticklabels(), visible=False);
#         plt.setp(axs.flatten()[1].get_yticklabels(), visible=False);
#         plt.setp(axs.flatten()[2].get_xticklabels(), visible=False);
#         plt.setp(axs.flatten()[2].get_yticklabels(), visible=False);
```
